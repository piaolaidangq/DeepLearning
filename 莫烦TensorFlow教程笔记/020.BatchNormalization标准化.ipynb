```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from keras import backend as K
K.clear_session()

# ACTIVATION = tf.nn.relu
ACTIVATION = tf.nn.tanh
N_LAYERS = 7
N_HIDDEN_UNITS = 30

def fix_seed(seed=1):                    # 设置随机数种子，使代码每次执行结果一致
    np.random.seed(seed)
    tf.set_random_seed(seed)
    
def plot_his(inputs, inputs_norm):       # 为每层神经网络的输入绘图
    for j, all_inputs in enumerate([inputs, inputs_norm]):
        for i, ainput in enumerate(all_inputs):
            plt.subplot(2, len(all_inputs), j*len(all_inputs)+(i+1))
            plt.cla()
            if i == 0:
                the_range = (-7, 10)
            else:
                the_range = (-1, 1)
            plt.hist(ainput.ravel(), bins=15, range=the_range, color='#FF5733')
            plt.yticks(())
            if j == 1:
                plt.xticks(the_range)
            else:
                plt.xticks(())
            ax = plt.gca()
            ax.spines['right'].set_color('none')
            ax.spines['top'].set_color('none')
        plt.title('%s normalizing' % ('Without' if j==0 else 'With'))
    plt.draw()
    plt.pause(0.01)
    
def built_net(xs, ys, norm):
    def add_layer(inputs, in_size, out_size, activation_function=None, norm=False):
        Weights = tf.Variable(tf.random_normal([in_size, out_size], mean=0., stddev=1.))
        biases = tf.Variable(tf.zeros([1, out_size])+0.1)
        Wx_plus_b = tf.matmul(inputs, Weights)+biases
        
        # 标准化（Batch Normalize）
        if norm:       
            # 计算均值和方差
            fc_mean, fc_var = tf.nn.moments(
                Wx_plus_b, 
                axes=[0],     # axes=[0]，设置标准化的维度，此处表示标准化 batch
                              # 对于图像，axes=[ 0,1,2]，表示标准化[batch,height, width]，无 channel
            )  
            scale = tf.Variable(tf.ones([out_size]))         # 扩大参数
            shift = tf.Variable(tf.zeros([out_size]))        # 平移参数
            epsilon = 0.001
            
#             # 批量训练时，均值和方差采用移动平均方式更新
#             ema = tf.train.ExponentialMovingAverage(decay=0.5)
#             def mean_var_with_update():
#                 ema_apply_op = ema.apply([fc_mean, fc_var])
#                 with tf.control_dependencies([ema_apply_op]):
#                     return tf.identity(fc_mean), tf.identity(fc_var)
#             mean, var = mean_var_with_update()
#             Wx_plus_b = tf.nn.batch_normalization(Wx_plus_b, mean, var, shift, scale, epsilon)

            Wx_plus_b = tf.nn.batch_normalization(Wx_plus_b, fc_mean, fc_var, shift, scale, epsilon)
            # 等同于
            # Wx_plus_b = (Wx_plus_b-fc_mean)/tf.sqrt(fc_var+0.001)     # 标准化
            # Wx_plus_b = Wx_plus_b*scale+shift                         # 扩大平移
            
        # 激活函数
        if activation_function is None:
            outputs = Wx_plus_b
        else:
            outputs = activation_function(Wx_plus_b)
            
        return outputs
    
    fix_seed(1)
    
    # 对第一次输入数据做 BN 处理
    if norm:
        fc_mean, fc_var = tf.nn.moments(
            xs, 
            axes=[0],
        )
        scale = tf.Variable(tf.ones([1]))
        shift = tf.Variable(tf.zeros([1]))
        epsilon = 0.001
        
#         # 批量训练时，均值和方差采用移动平均方式更新
#         ema = tf.train.ExponentialMovingAverage(decay=0.5)
#         def mean_var_with_update():
#             ema_apply_op = ema_apply([fc_mean, fc_var])
#             with tf.control_dependencies([ema_apply_op]):
#                 return tf.identity(fc_mean), tf.identity(fc_var)
#         mean, var = mean_var_with_update()
#         xs = tf.nn.batch_normalization(xs, mean, var, shift, scale, epsilon)

        xs = tf.nn.batch_normalization(xs, fc_mean, fc_var, shift, scale, epsilon)
        
    # 记录每层输入（绘图使用）
    layers_inputs = [xs]
    
    # 构建隐藏层
    for l_n in range(N_LAYERS):
        layer_input = layers_inputs[l_n]
        in_size = layers_inputs[l_n].get_shape()[1].value
        
        output = add_layer(
            layer_input,     # input
            in_size,         # input size
            N_HIDDEN_UNITS,  # output size
            ACTIVATION,      # activation function
            norm,            # normalize before activation
        )
        layers_inputs.append(output)        # 记录 output，即下一层的输入（绘图使用）
        
    # 构建输出层
    prediction = add_layer(layers_inputs[-1], 30, 1, activation_function=None)
    
    cost = tf.reduce_mean(tf.reduce_sum(tf.square(ys-prediction), reduction_indices=[1]))
    train_op = tf.train.GradientDescentOptimizer(0.001).minimize(cost)
    return [train_op, cost, layers_inputs]

# 制作数据
fix_seed(1)
# x_data = np.linspace(-7, 10, 2500)[:, np.newaxis]
x_data = np.linspace(-7, 10, 500)[:, np.newaxis]
# np.random.shuffle(x_data)
noise = np.random.normal(0, 8, x_data.shape)
y_data = np.square(x_data) - 5 + noise

# 制作数据绘图
# plt.scatter(x_data, y_data)
# plt.show()

xs = tf.placeholder(tf.float32, [None, 1])
ys = tf.placeholder(tf.float32, [None, 1])

# 构建无 BN 和有 NB 的神经网络做对比
train_op, cost, layers_inputs = built_net(xs, ys, norm=False)                  # 无 BN 的神经网络
train_op_norm, cost_norm, layers_inputs_norm = built_net(xs, ys, norm=True)    # 有 BN 的神经网络

sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 记录有 BN 和无 NB 的神经网络的 cost
cost_his = []
cost_his_norm = []
record_step = 5

plt.ion()
plt.figure(figsize=(7, 3))
for i in range(250):
    if i%50 == 0:
        # 绘制输入数据分布的直方图
        all_inputs, all_inputs_norm = sess.run([layers_inputs, layers_inputs_norm], feed_dict={xs: x_data, ys:y_data})
        plot_his(all_inputs, all_inputs_norm)
        
    sess.run([train_op, train_op_norm], feed_dict={xs: x_data, ys: y_data})
    
    if i%record_step == 0:
        # 每 5 步记录 cost
        cost_his.append(sess.run(cost, feed_dict={xs: x_data, ys: y_data}))
        cost_his_norm.append(sess.run(cost_norm, feed_dict={xs: x_data, ys: y_data}))
        
# 绘制 cost（误差值）图
plt.ioff()
plt.figure()
plt.plot(np.arange(len(cost_his))*record_step, np.array(cost_his), label='no BN')
plt.plot(np.arange(len(cost_his_norm))*record_step, np.array(cost_his_norm), label='BN')
plt.legend()
plt.show()
```

### tf.identity()，赋值操作。
* tf.identity在计算图内部创建了两个节点，send / recv节点，用来发送和接受两个变量；
* 若两个变量在不同的设备上，则执行赋值变量操作
* 若在相同设备上，则执行引用操作。

**注意**：与 y = x 区别

```python
x = tf.Variable(0., dtype=tf.float32)
x_plus_1 = tf.assign_add(x, 1.)           # x = x+1 操作

with tf.control_dependencies([x_plus_1]): 
    y = x                                 # 错误赋值操作                               
    
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    print(sess.run(y))
```

**分析**：
* tf.control_dependencies([a,]):b

    执行 b 之前需要执行 a
* y 结果为 0.0，并没有赋值成功，原因为：虽然执行了 x=x+1 操作，但 y = x 并不在计算图中
* 需要使用 tf.identity() 进行赋值，将该操作建立在计算图中

**正确示例如下**：

```python
x = tf.Variable(0., dtype=tf.float32)
x_plus_1 = tf.assign_add(x, 1.)           # x = x+1 操作

with tf.control_dependencies([x_plus_1]): 
    y = tf.identity(x)                    # 正确赋值操作                                                
    
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    print(sess.run(y))
```

### tf.control_dependencies()，flow顺序控制机制，该函数保证其辖域中的操作必须要在该函数所传递的参数中的操作完成后再进行

**示例**：

```python
x = tf.Variable(0., dtype=tf.float32)
x_plus_1 = tf.assign_add(x, 1.)           # x = x+1 操作

with tf.control_dependencies([x_plus_1]):
    y = tf.identity(x)                    # 执行 y 时需要先执行 x_plus_1
    
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for i in range(5):
        print(sess.run(y))
```

### tf.assign_add(ref, value, use_locking=None, name=None)，将值 value 加到变量 ref 上， 即 ref = ref +value。
* ref，变量
* value，值
* use_locking，默认 False, 若为 True，则受锁保护
* name，名称

```python
import tensorflow as tf

x = tf.Variable(1., dtype=tf.float32)
y = tf.constant(2., dtype=tf.float32)
z = tf.assign_add(x, y)                      # 单个值相加

X = tf.Variable(tf.ones([2, 3]), dtype=tf.float32)
Y = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
Z = tf.assign_add(X, Y)                      # 矩阵对应位置相加
init = tf.initialize_all_variables()

with tf.Session() as sess:
    sess.run(init)
    print('加操作前 x:', sess.run(x))
    print('加操作 z:', sess.run(z))
    print('加操作后 x:', sess.run(x))
    
    print('='*30)
    
    print('加操作前 X:\n', sess.run(X))
    print('加操作 Z:\n', sess.run(Z))
    print('加操作后 X:\n', sess.run(X))
```

### tf.assign_sub(ref, value, use_locking=None, name=None)，变量 ref 减去 value值，即 ref = ref - value。
* ref，变量
* value，值
* use_locking，默认 False, 若为 True，则受锁保护
* name，名称

```python
import tensorflow as tf

x = tf.Variable(1., dtype=tf.float32)
y = tf.constant(2., dtype=tf.float32)
z = tf.assign_sub(x, y)                      # 单个值相减

X = tf.Variable(tf.ones([2, 3]), dtype=tf.float32)
Y = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
Z = tf.assign_sub(X, Y)                      # 矩阵对应位置相减
init = tf.initialize_all_variables()

with tf.Session() as sess:
    sess.run(init)
    print('减操作前 x:', sess.run(x))
    print('减操作 z:', sess.run(z))
    print('减操作后 x:', sess.run(x))
    
    print('='*30)
    
    print('减操作前 X:\n', sess.run(X))
    print('减操作 Z:\n', sess.run(Z))
    print('减操作后 X:\n', sess.run(X))
```

### tf.assign(ref, value, validate_shape=True, use_locking=None, name=None)，将　value 赋值给　ref，即 ref = value
* ref，变量
* value，值
* validate_shape，默认 True，值 shape 需要与变量 shape 一致；若设为 False，则值 shape 覆盖变量 shape
* use_locking，默认 False，若为 True，则受锁保护
* name，名称

```python
import tensorflow as tf

x = tf.Variable(1., dtype=tf.float32)
y = tf.constant(2., dtype=tf.float32)
z = tf.assign(x, y)                             # 单个值赋值

X = tf.Variable(tf.ones([2, 3]), dtype=tf.float32)
Y = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
Z = tf.assign(X, Y)                              # 矩阵赋值（变量与值 shape 相同）

X_ = tf.Variable(1.0, dtype=tf.float32)
Y_ = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
Z_ = tf.assign(X_, Y_, validate_shape=False)    # 矩阵赋值（变量与值 shape 不同）

init = tf.initialize_all_variables()

with tf.Session() as sess:
    sess.run(init)
    print('赋值操作前 x:', sess.run(x))
    print('赋值操作 z:', sess.run(z))
    print('赋值操作后 x:', sess.run(x))
    
    print('='*30)
    
    print('赋值操作前 X:\n', sess.run(X))
    print('赋值操作 Z:\n', sess.run(Z))
    print('赋值操作后 X:\n', sess.run(X))
    
    print('='*30)
    
    print('赋值操作前 X_:\n', sess.run(X_))
    print('赋值操作 Z_:\n', sess.run(Z_))
    print('赋值操作后 X_:\n', sess.run(X_))
```

### tf.pow(x, y, name=None)，幂值计算操作，即 x^y。
* 同 tf.math.pow(x, y, name=None)

```python
import tensorflow as tf

x = tf.constant(2, dtype=tf.float32)
y = tf.constant(3, dtype=tf.float32)
z = tf.pow(x, y)                              # 标量^标量操作

X1 = tf.constant(3, dtype=tf.float32)
Y1 = tf.constant([1, 2, 3], dtype=tf.float32)        
Z1 = tf.pow(X1, Y1)                           # 标量^向量操作（矩阵同理）

X2 = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
Y2 = tf.constant([[2,2,2],[3,3,3]], dtype=tf.float32)
Z2 = tf.pow(X2, Y2)                           # 矩阵^矩阵操作（向量同理），对应位置的值进行幂值计算

with tf.Session() as sess:
    print(sess.run(z))
    print(sess.run(Z1))
    print(sess.run(Z2))
```
